## Part 5: Filtering Rows (Where)


### 5.1 Basic Filtering

**Berkeley datascience way:**
```python
nba.where('SALARY', are.above(10))
```

**Pandas way:** Use boolean indexing

````python
# Find players earning more than $10 million
high_earners = nba[nba['SALARY'] > 10]
print(f"Number of players earning > $10M: {len(high_earners)}")
high_earners.head(10)
````

````python
# How does boolean indexing work? Let's break it down:
condition = nba['SALARY'] > 10
print("The condition creates a Series of True/False:")
print(condition.head(10))
print(f"\nType: {type(condition)}")
````

**Key Concept:** Boolean indexing works in two steps:
1. Creates a condition that returns True/False for each row
2. Uses that to filter: `df[condition]`

### 5.2 Comparison Operators

| Operator | Meaning | Example |
|----------|---------|----------|
| `==` | Equal to | `df['col'] == 5` |
| `!=` | Not equal to | `df['col'] != 5` |
| `>` | Greater than | `df['col'] > 5` |
| `>=` | Greater than or equal | `df['col'] >= 5` |
| `<` | Less than | `df['col'] < 5` |
| `<=` | Less than or equal | `df['col'] <= 5` |

````python
# Find Stephen Curry
curry = nba[nba['PLAYER'] == 'Stephen Curry']
print("Stephen Curry's information:")
curry
````

````python
# Find all Golden State Warriors
warriors = nba[nba['TEAM'] == 'Golden State Warriors']
print(f"Number of Warriors players: {len(warriors)}")
warriors.sort_values('SALARY', ascending=False)
````

### 5.3 Multiple Conditions

Use `&` for AND, `|` for OR (must use parentheses!)

````python
# Point Guards earning more than $15 million
rich_point_guards = nba[(nba['POSITION'] == 'PG') & (nba['SALARY'] > 15)]
print("Point Guards earning > $15M:")
rich_point_guards.sort_values('SALARY', ascending=False)
````

````python
# Players earning between $10M and $10.3M
mid_salary = nba[(nba['SALARY'] >= 10) & (nba['SALARY'] < 10.3)]
mid_salary.sort_values('SALARY')
````

**Common Mistake:** Remember to use `&` (and) or `|` (or), NOT `and` or `or`!

## Part 7: Real Data Analysis - US Census

### 7.1 Loading Census Data

````python
# Load census data
census_url = 'http://www2.census.gov/programs-surveys/popest/technical-documentation/file-layouts/2010-2019/nc-est2019-agesex-res.csv'
full_census = pd.read_csv(census_url)

print(f"Shape: {full_census.shape}")
print(f"\nColumns: {full_census.columns.tolist()}")
full_census.head()
````

### 7.2 Data Cleaning and Selection

````python
# Select only the columns we need
us_pop = full_census[['SEX', 'AGE', 'POPESTIMATE2014', 'POPESTIMATE2019']].copy()

# Rename for simplicity
us_pop = us_pop.rename(columns={
    'POPESTIMATE2014': '2014',
    'POPESTIMATE2019': '2019'
})

print("Cleaned census data:")
us_pop.head(10)
````

**Understanding the Data:**
- `SEX`: 0 = Total, 1 = Male, 2 = Female
- `AGE`: Age in years (999 = all ages, 100 = 100+)
- `2014`, `2019`: Population estimates

### 7.3 Filtering for Total Population by Age

````python
# Get total population (SEX = 0) and drop the SEX column
us_pop_by_age = us_pop[us_pop['SEX'] == 0].drop(columns=['SEX']).copy()

# Remove the summary row (AGE = 999)
us_pop_by_age = us_pop_by_age[us_pop_by_age['AGE'] != 999]

print("Population by age:")
us_pop_by_age.head(10)
````

### 7.4 Calculating Population Change

````python
# Calculate absolute and percent change
us_pop_by_age['Change'] = us_pop_by_age['2019'] - us_pop_by_age['2014']
us_pop_by_age['Percent Change'] = (us_pop_by_age['Change'] / us_pop_by_age['2014']) * 100

# Round for readability
us_pop_by_age['Percent Change'] = us_pop_by_age['Percent Change'].round(2)

print("Population changes:")
us_pop_by_age.head(10)
````

### 7.5 Finding the Biggest Changes

````python
# Sort by absolute change (descending)
biggest_changes = us_pop_by_age.sort_values('Change', ascending=False)

print("Top 10 ages with largest population increases:")
biggest_changes.head(10)
````

**Key Insight:** Ages 68-72 saw the largest increases! Why?

These people were born 1947-1951 - the **post-WWII baby boom**!

### 7.6 Visualization

````python
# Create a bar chart of the top 15 age groups by change
top_15 = biggest_changes.head(15)

plt.figure(figsize=(12, 6))
plt.bar(top_15['AGE'], top_15['Change'] / 1000000, color='steelblue')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Population Change (millions)', fontsize=12)
plt.title('Top 15 Ages by Population Increase (2014-2019)', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````

## Part 8: Sex Ratios Analysis

### 8.1 Separating Male and Female Data

````python
# Get 2019 data only
us_pop_2019 = us_pop[['SEX', 'AGE', '2019']].copy()

# Separate males and females (excluding age 999)
females = us_pop_2019[(us_pop_2019['SEX'] == 2) & (us_pop_2019['AGE'] != 999)].copy()
males = us_pop_2019[(us_pop_2019['SEX'] == 1) & (us_pop_2019['AGE'] != 999)].copy()

print(f"Females shape: {females.shape}")
print(f"Males shape: {males.shape}")
````

### 8.2 Calculating Sex Ratios

````python
# Create a new DataFrame with ratios
# We need to reset the index to make sure the division works correctly
females_reset = females.reset_index(drop=True)
males_reset = males.reset_index(drop=True)

ratios = pd.DataFrame({
    'AGE': females_reset['AGE'],
    'Female Population': females_reset['2019'],
    'Male Population': males_reset['2019'],
    'F:M Ratio': females_reset['2019'] / males_reset['2019']
})

ratios['F:M Ratio'] = ratios['F:M Ratio'].round(3)

print("Sex ratios by age:")
ratios.head(10)
````

**Understanding the Ratio:**
- Ratio < 1.0: More males than females
- Ratio = 1.0: Equal numbers
- Ratio > 1.0: More females than males

### 8.3 Analyzing Young Ages

````python
# Infants (age 0)
infant_ratio = ratios[ratios['AGE'] == 0]
print("Infant (age 0) sex ratio:")
print(infant_ratio)
print(f"\nThis means there are about {infant_ratio['F:M Ratio'].iloc[0]:.3f} female infants for every male infant")
print(f"In other words, about {(1 - infant_ratio['F:M Ratio'].iloc[0]) * 100:.1f}% more male infants than female")
````

### 8.4 Analyzing Older Ages

````python
# Senior citizens (75+)
seniors = ratios[ratios['AGE'] > 75]
print("Sex ratios for ages 75+:")
seniors
````

````python
# Age 90+
very_old = ratios[ratios['AGE'] >= 90]
print("Sex ratios for ages 90+:")
print(very_old)
````

**ðŸ’¡ Key Insight:** At age 99, there are about 2.9 females for every male!

### 8.5 Visualizing Sex Ratios

````python
# Create a line plot
plt.figure(figsize=(14, 6))
plt.plot(ratios['AGE'], ratios['F:M Ratio'], linewidth=2, color='steelblue')
plt.axhline(y=1.0, color='red', linestyle='--', linewidth=1, alpha=0.7, label='Equal ratio')

plt.xlabel('Age', fontsize=12)
plt.ylabel('Female to Male Ratio', fontsize=12)
plt.title('Female to Male Population Ratio by Age (2019)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()

# Add annotation
plt.annotate('Baby boom generation\n(ages 68-72)',
             xy=(70, ratios[ratios['AGE']==70]['F:M Ratio'].iloc[0]),
             xytext=(50, 1.5),
             arrowprops=dict(arrowstyle='->', color='black', alpha=0.7),
             fontsize=10)

plt.tight_layout()
plt.show()
````

````python
# Find the age where ratio first exceeds 1.0
crossover = ratios[ratios['F:M Ratio'] > 1.0].iloc[0]
print(f"The ratio first exceeds 1.0 at age {crossover['AGE']}")
print(f"At that age, the ratio is {crossover['F:M Ratio']:.3f}")
````
