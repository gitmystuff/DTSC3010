## PEMDAS / BODMAS

* Parentheses/Brackets: Operations inside grouping symbols are performed first.
* Exponents/Orders: Exponents (powers) are evaluated next.
* Multiplication and Division: These operations are performed from left to right.
* Addition and Subtraction: These operations are performed last, from left to right.

In the BODMAS acronym, Division (D) appears before Multiplication (M) solely for mnemonic convenience; however, neither has priority over the other.

In mathematics, division and multiplication are considered inverse operations at the same rank or level of precedence.

**The Real Rule: Left-to-Right**

When an expression contains both multiplication and division, you must solve them in the order they appear from left to right.

* BODMAS (Division then Multiplication) and PEMDAS Multiplication then Division) are both correct only if you follow the left-to-right sub-rule.
* If you strictly followed the acronym letters without the left-to-right rule, you would often get different and incorrect answers.

**Why They Are Equal**

* **Inverse Operations:** Division is mathematically defined as multiplication by a reciprocal (e.g., \(10\div 2\) is the same as \(10\times \frac{1}{2}\)).
* **Axiomatic Priority:** Because they are essentially the same operation, they share the same priority level.
* **Consistency:** Standardizing this "equal rank" ensures that everyone arrives at the same result regardless of which mnemonic (BODMAS, PEMDAS, etc.) they use.

## Names

In the context of Python, **names** and **variables** refer to the exact same thing.

While most programming languages and general computer science resources use the term "variable," Python resources (like our book) often prefer the term "name" to be technically precise about how Python works under the hood.

Here is the breakdown of why they are the same, yet distinct:

### 1. The Common Term: "Variable"

In almost any conversation, tutorial, or job setting, you will hear people say **"assign a value to a variable."**

* **Concept:** A container that holds data.
* **Usage:** Universal across programming languages (C++, Java, etc.).

### 2. The Pythonic Term: "Name"

Python technically doesn't have "variables" in the traditional sense of a "box" that holds a value. Instead, Python has **Names** that act like **labels** or tags attached to objects in memory.

* **Concept:** A name tag tied to a specific object.
* **The Difference:** When you say `x = 10`, you aren't putting the number 10 inside a box labeled `x`. You are creating a number object `10` and tying a name tag labeled `x` onto it.

### Summary Comparison

| Term | Context | How to visualize it |
| --- | --- | --- |
| **Variable** | General programming | A **box** that you put a value inside. |
| **Name** | Python internals | A **sticky note** you stick onto a value. |

**The Bottom Line:**

## What is Python?

* **Interpreted**: a computer program that directly executes instructions written in a programming or scripting language, without requiring them previously to have been compiled into a machine language program. (Compiled languages: Go, C++)
* **Object Oriented**: a programming paradigm based on the concept of "objects", which can contain data and code: data in the form of fields, and code, in the form of procedures. A common feature of objects is that procedures are attached to them and can access and modify the object's data fields.
* **Object**: An object is simply a collection of data (variables) and methods (functions) that act on those data. Similarly, a class is a blueprint for that object. We can think of a class as a sketch (prototype) of a house. It contains all the details about the floors, doors, windows, etc.
* **High-Level Language**: A high-level language (HLL) is a programming language such as C, FORTRAN, or Pascal that enables a programmer to write programs that are more or less independent of a particular type of computer. Such languages are considered high-level because they are closer to human languages and further from machine languages. Python, C#, C++, PHP, Java vs assembly language and machine code
* **Dynamic Semantics**: dynamic objects are instances of values contained into constructs in the code, and they exist at run-time level. Furthermore, we can assign to one object multiple values, since it will update itself, differently from a static semantic language. Namely, if we set a=2 and then a=’hello’, the string value will substitute the integer one as soon as the line is executed
* **Built-In Data Structures**: Organizing, managing, and storing data is important as it enables easier access and efficient modifications. Data Structures allows you to organize your data in such a way that enables you to store collections of data, relate them and perform operations on them accordingly. Python has implicit support for Data Structures which enable you to store and access data. Some of these structures are called List, Dictionary, Tuple and Set.
* **Dynamic Typing**: The term dynamic typing means that a compiler or an interpreter assigns a type to all the variables at run-time. The type of a variable is decided based on its value. The programs written using dynamic-typed languages are more flexible but will compile even if they contain errors.


Performance Trade-Off

* **Static Typing**: Because types are checked at compile time, the program knows exactly what kind of data is in each variable before it runs. This allows the compiler to create highly optimized machine code that can execute very efficiently. There is no need for extra checks during runtime, which saves processing time.
* **Dynamic Typing**: With dynamic typing, the program must perform type checks on the fly, as the code is being executed. This adds a small overhead to every operation, as the program has to determine the type of the variable at that exact moment. This continuous process of checking types at runtime makes dynamically typed languages inherently slower than their statically typed counterparts.

While this performance difference exists, it's often not a major issue for most applications. The speed and flexibility of dynamic languages like Python make them ideal for many tasks, such as scripting, web development, and data analysis, where development speed is more important than raw execution speed. For applications where performance is critical (like system software, game engines, or high-frequency trading), a statically typed language is usually the better choice.

## Objects

To understand Python objects and classes, it helps to think of them as the **blueprint** versus the **actual house**.

In Python, everything is an object. A **Class** is the template that defines what that object will look like and what it can do.

---

### 1. The Class (The Blueprint)

A class is a logical grouping of data and functions. It defines two things:

* **Properties (Attributes):** What the object "is" or "has."
* **Behaviors (Methods):** What the object "does."

### 2. The Object (The Instance)

An object is a specific "copy" of the class with real data. While the blueprint says "a car has a color," the object says "this specific car is **Red**."

### 3. Let's Build a Class

Using your interest in the NBA, let’s look at how a `Player` class would look compared to a specific player object.

```python
class NBAPlayer:
    def __init__(self, name, team, points):
        # Attributes: Stored data
        self.name = name
        self.team = team
        self.points = points

    # Method: An action the object performs
    def celebrate(self):
        return f"{self.name} does a signature dance!"

# Creating Objects (Instances)
player1 = NBAPlayer("LeBron", "Lakers", 25)
player2 = NBAPlayer("Steph", "Warriors", 30)

# Accessing an attribute (No parentheses)
print(player1.team)  # Output: Lakers

# Calling a method (Requires parentheses)
print(player2.celebrate()) # Output: Steph does a signature dance!

```

---

### Key Takeaways

* **Classes** are the "idea" of a thing (the `NBAPlayer` code).
* **Objects** are the "actual" things (`player1`, `player2`).
* **`self`**: You'll see `self` inside class methods. It is just a way for the code to say "apply this action to **this specific** object."

---

### 1. The Attribute: `nba.shape`

An **attribute** is a variable stored inside an object. It represents a "state" or a piece of information about the object.

* **Syntax:** No parentheses `()` are used.
* **What it does:** It simply retrieves a stored value. In this case, `.shape` tells you the dimensions (rows, columns) of the `nba` dataset.
* **Analogy:** If `nba` were a person, an attribute would be `nba.height`. You aren't "doing" anything; you're just checking a measurement.

### 2. The Method: `nba.info()`

A **method** is a function that belongs to an object. It represents an "action" or a behavior that the object can perform.

* **Syntax:** Requires parentheses `()`, even if they are empty.
* **What it does:** It performs a process. When you call `.info()`, the computer calculates and prints a summary of the DataFrame, including data types and memory usage.
* **Analogy:** Using the person example, a method would be `nba.speak()`. It is an action that requires the object to "run" some code.

## Zero Indexing

## Interval Notation

The notation tells you whether the endpoints of the range are included or excluded.

A square bracket [ or ] means the endpoint is included in the interval.

A parenthesis ( or ) means the endpoint is excluded from the interval.

So, let's break down the notation [0, 5):

The square bracket [ next to 0 means that 0 is included in the range.

The parenthesis ) next to 5 means that 5 is excluded from the range.

Therefore, the set of numbers represented by [0, 5) includes 0, and all real numbers up to, but not including, 5. This would be the correct way to write it. The notation [0, 1, 2, 3, 4) is incorrect because interval notation represents a continuous range of numbers, not a discrete list of integers.

````python
# https://docs.python.org/3/library/fractions.html
from fractions import Fraction

a = 1/6
print(a, type(a))
a = Fraction(a) # .limit_denominator(1000)
print(a, type(a))
````

The `.limit_denominator()` method shows the difference between how a **Primitive Type** (float) and a **Non-Primitive Class** (Fraction) handle the same number.

**1. The Float Phase (The "Messy" Primitive)**

* **Code:** `a = 1/6`
* **Result:** `0.16666666666666666 <class 'float'>`
* **Explanation:** When you define `a` as `1/6`, Python stores it as a `float`. Because computers use binary (base-2) to store decimals, they cannot store  perfectly. It ends up being an approximation that is slightly "off" at the very end.

**2. The Fraction Phase (The Precise Class)**

* **Code:** `a = Fraction(a)`
* **Result:** `6004799503160661/3602879018963968 <class 'fractions.Fraction'>`
* **Explanation:** Here, you converted the float into a **Fraction object**.
* **The Problem:** Because the input (`0.1666...`) was already an imperfect approximation, the `Fraction` class tried to be **too literal**. It created a massive, ugly fraction that mathematically equals that exact messy decimal you gave it.

This is exactly why you need `.limit_denominator()`. If you add that method to your code, it tells the object: *"I know the math looks like that giant fraction, but find me the simplest version that is close to it."*

# Gottfried Wilhelm Leibniz

Gottfried Wilhelm Leibniz was perhaps the ultimate polymath—a mathematician, diplomat, and philosopher who sought to synthesize all human knowledge into a single, logical system.

His contributions to the topics you mentioned often stem from a core desire for harmony and the **Principle of Sufficient Reason**: the idea that nothing happens without a reason why it should be so rather than otherwise.

---

## Bias and Beliefs

In his *New Essays on Human Understanding*, Leibniz addressed how humans form beliefs and the "pre-judices" that lead to error.

* **Innate Ideas:** Unlike Locke (who argued for a *tabula rasa*), Leibniz believed the mind is like a "veined block of marble." The veins represent innate tendencies or "dispositions" of thought.
* **Petites Perceptions:** He theorized that we have many unconscious "minute perceptions" (unconscious biases) that influence our conscious beliefs.
* **The Solution:** He hoped to eliminate human bias through his **Characteristica Universalis**—a universal symbolic language. He famously said that if two people disagreed, they should simply sit down with their pens and say, *"Calculemus"* (Let us calculate), to find the objective truth.

---

## Determinism vs. Probability

Leibniz was a pioneer in both the metaphysics of necessity and the mathematics of chance.

* **Determinism:** Leibniz held a view called **Predicate-in-Notion**. This means that the concept of an individual (like "Cliff") contains everything that will ever happen to them. From God's perspective, the future is "encoded" in the present.
* **Probability:** Despite this "hard" logical determinism, Leibniz was one of the first to treat **probability** as a formal logic for uncertain situations. He applied this to law and insurance, arguing that since humans lack "infinite analysis" (we can't see the full code), we must use probability to quantify our ignorance.

---

## Class and Object

While "Class" and "Object" are modern programming terms, Leibniz’s **Monadology** is often considered a spiritual precursor to Object-Oriented Programming (OOP).

* **The Monad as an Object:** Leibniz's "Monads" are the building blocks of reality. They are like software objects:
* **Encapsulation:** He said Monads "have no windows." They do not interact with each other; their "state" is entirely internal.
* **State and Logic:** Every Monad has *perception* (current state) and *appetition* (the internal logic/drive to move to the next state).


* **Classes:** In his work on logical calculus, he treated properties as sets. He worked on a "calculus of identities" where complex concepts (the Class) are built from the intersection of simpler "primitive" attributes.

---

## Mind and Matter (Pre-established Harmony)

Leibniz rejected the idea that the mind and body "push" each other (interactionism) or that God intervenes every second (occasionalism).

Instead, he proposed **Pre-established Harmony**. Imagine two clocks that keep perfect time. They aren't connected, but when one strikes twelve, the other does too because they were both synchronized by a master clockmaker at the start.

* **The Mind** follows its own laws of final causes (desires/logic).
* **The Body** follows its own laws of efficient causes (physics).
* They appear to interact only because God "programmed" them to be perfectly synchronized from the beginning.

---

## Aristotle: Form and Matter

Leibniz performed a "rehabilitation" of Aristotelian concepts within the new mechanical science of the 17th century.

* **Substantial Form:** Aristotle believed every object has a "form" that makes it what it is. Early modern scientists (like Descartes) rejected this as "occult."
* **Leibniz’s Twist:** He argued that we *need* substantial forms to explain **unity**. A pile of rocks is just a collection, but a living thing has a "form" (which he called an *Entelechy* or soul) that holds it together.
* **Active vs. Passive Force:** * **Form** = *Primitive Active Force* (the "spark" or internal energy).
* **Matter** = *Primitive Passive Force* (impenetrability and resistance to motion).

---

## Calculus vs. Newton

The "Calculus Wars" was one of the most famous disputes in history. Both men discovered calculus independently, but their approaches were different:

* **Newton’s Fluxions:** Newton arrived at calculus through physics (kinematics). He thought of variables as "flowing" over time.
* **Leibniz’s Differentials:** Leibniz arrived through logic and geometry. He focused on the sum of infinitesimal differences.
* **The Winner (Notationally):** We use Leibniz's notation today because it is far more flexible. He gave us the "integral" symbol  (a stylized 'S' for *summa*) and the "d" notation (like ), whereas Newton's "dot notation" () is mostly relegated to specific physics contexts.

## Space and Time vs. Newton

Leibniz and Newton held fundamentally different views on the "fabric" of reality, a debate famously captured in the *Leibniz-Clarke Correspondence*.

* **Newton (Absolutism):** Viewed space and time as absolute "containers" created by God. Even if you removed all matter, space and time would still exist as an empty stage.
* **Leibniz (Relationalism):** Argued that space and time are merely **relations** between objects.
* **Space** is the "order of co-existing things." Without objects, "space" is nothing.
* **Time** is the "order of successive things." Without events, "time" is nothing.
* **The Argument:** Leibniz used the *Identity of Indiscernibles* to argue that if God moved the entire universe two feet to the left, and everything stayed the same relative to each other, there would be no reason for God to prefer one position over the other. Therefore, absolute space cannot exist because it would violate the Principle of Sufficient Reason.
